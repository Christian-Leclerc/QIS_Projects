```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
response = requests.get(
	url="https://www.edreams.com/blog/in-flight-wifi/",
)
print(response.status_code)
```

```python
body = response.text
soup = BeautifulSoup(response.content, "html.parser")
print(soup.prettify())
```

```python
 # Get all the airline lists
results = soup.findAll('div', attrs={'class': 'su-spoiler su-spoiler-style-fancy su-spoiler-icon-plus su-spoiler-closed'})
wifi_list = [] # Set up list for all values
for i, j in enumerate(results):
    bullet_dict = dict() # Dict tp place values in
    aircraft_name = j.findAll('div', attrs={'class': 'su-spoiler-title'})
    bullet_dict['Airline'] = aircraft_name[0].text # Get the name of the airline
    content = j.findAll('div', attrs={'class': 'su-spoiler-content su-u-clearfix su-u-trim'})
    for y in content:
        li = y.findAll('li')
        for x in li:
            text = x.get_text()
            text_list = text.split(':') # The category and values are already separated by ":"
            if text_list[0] == "Providers":
                text_list[0] = "Provider" # Standardize the category "Provider"
            bullet_dict[text_list[0]] = text_list[1]
    wifi_list.append(bullet_dict) # Add dictionary values into list
df = pd.DataFrame(wifi_list)
df
```

```python
lists = soup.findAll("ul") # Get all the unorganized lists
wifi_list = [] # Set up list for all values
for i, j in enumerate(lists):
    if i >= 6 and i < 60: # Locate where are the airline data is
        li = j.findAll('li') # Get all the lists within the unorganized lists
        bullet_dict = dict() # Dict tp place values in
        for x in li:
            text = x.get_text()
            text_list = text.split(':') # The category and values are already separated by ":"
            if text_list[0] == "Providers":
                text_list[0] = "Provider" # Standardize the category "Provider"
            bullet_dict[text_list[0]] = text_list[1]
        wifi_list.append(bullet_dict)
df = pd.DataFrame(wifi_list)
df
```
